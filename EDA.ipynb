```python
# prompt: dame la caja de bigotes de Area_parqeo

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'Area_parqeo' is the column name for the parking area
# If the column name is different, replace 'Area_parqeo' with the actual column name

# Check if 'Area_parqeo' column exists
if 'Area_parqeo' in df.columns:
    plt.figure(figsize=(8, 6))
    df.boxplot(column=['Area_parqeo'])
    plt.title('Boxplot of Parking Area')
    plt.ylabel('Area_parqeo')
    plt.show()

    # Calculate and print quartiles and other statistics
    area_parqeo_stats = df['Area_parqeo'].describe()
    print(area_parqeo_stats)
else:
    print("Column 'Area_parqeo' not found in the DataFrame.")
```

```python
# prompt: borrar todos los registros que sean superior a 32 y dejarlo en el mismo df

# Assuming 'habitaciones' is the column you want to filter
df = df[df['Area_parqeo'] <= 32]
if 'Area_parqeo' in df.columns:
    plt.figure(figsize=(8, 6))
    df.boxplot(column=['Area_parqeo'])
    plt.title('Boxplot of Parking Area')
    plt.ylabel('Area_parqeo')
    plt.show()

    # Calculate and print quartiles and other statistics
    area_parqeo_stats = df['Area_parqeo'].describe()
    print(area_parqeo_stats)
else:
    print("Column 'Area_parqeo' not found in the DataFrame.")
```
